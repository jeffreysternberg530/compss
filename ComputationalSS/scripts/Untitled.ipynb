```python
data = open("../data/City List.csv", encoding='utf-8').read()
city = data.split()
```

```python
def anydup(thelist):
  seen = set()
  for x in thelist:
    if x in seen: print(x, True)
    seen.add(x)
  return False
```

```python
anydup(city)
```

```python
count = 0
for x in city:
    if x == "Reykjavik": 
        count = count+1
```
