### The City Image: From Mental Map to Marketing Tool 

The concept of the city image was first put forth by Kevin Lynch in his book *The Image of the City* (1960). Lynch, an urban planner and designer, defined the city image as "the 'public images,' the common mental pictures carried by large numbers of a city's inhabitants: areas of agreement which might be expected to appear in the interaction of a single physical reality, a common culture, and a basic physiological nature" (pg.7). For Lynch, the city image is the symbolic collection of each individual's experiences and associations with the city, whose clarity orients and makes possible the use of the city by large amounts of people for a variety of different purposes. Lynch saw it as the job of the city planner to work to clarify the city image, the shared and agreed upon mental map of the city, in order to make the city inhabital and useable by the largest collection of people. He saw the construction of the city image as an on-going, open-ended process whose shape is changed by each new experience of the people interacting with the built environment, and every new addition to the built environment prompting new interactions from the citizens of the city. 

Scholars of political economy (Harvey 1989, 1992; Logan and Molotch 1987; Logan 1976; Molotch 1976)  have taken the concept of the city image and put it center-stage to the actions of the neo-liberal city. Harvey Molotch, in *The City as Growth Machine* (1976) sees the construction of an image of the city as being one of the chief sites of strategy and importance for city elites. City elites attempt to control what city image is being broadcasted by their control and alliance with city newspapers and media institutions, to strategically attract capital and oppurtunities for growth from outside the city. David Harvey (1989, 1992) identifies this trend as emerging in the economic and strategic shift in cities from managerialism of Keynesianism to the entrepreneurialism of Neo-Liberalism. For Harvey, once national tax revenues ceased to be redistributed around the country, cities had to make up the gap by becoming entrepreneurs, by attracting industry and talent to support its operating revenues. It is in this phase that the city image shifts from being an organically produced subject of city planning, to a tool for entrepreneurialism. The successful marketing of a city becomes tantamount in ensuring the future survival and success of a city in the long run. 

The city image, once placed within city marketing schemes sparked by competition between cities for industry and talent, becomes a cultural object whose content is censored, curated, and produced by a small segment of the city's population, the growth coalition elite. The image constructed is no longer on the product of all the inhabitants of the city, but a fractured series of symbolic and social networks (Sevin 2014) utilized differently by different groups. The emergence of the creative city (Florida 2001, 2003) and the global city (Sassen 2001) provided prominent, globally circulated schema and sets of criteria. City governments and growth coalitions attempted to construct their city images, through both urban planning and media circuation, to become legible under the banners of the creative and/or global city. These images are circulated in order to attract populations that do not live there, to create new sites of orientation which offer up the city as a site of desire, of wish fulfillment, among them the cretive class and global expatriates and elites (Vanolo 2015). City images become a key mechanism in converting the use value of place into exchange value, into commodifying the urban space of the city into a consumable commodity (Lefebvre 1991). 

### Accesing the Authentic, Congealed City Image

These city images are both the result of, and influenced by cities' long social, cultural, and political-economic histories. These histories both congeal into a form of city character which leads cities down somewhat path-dependent urban development trajectories, as well as become the raw material from which new and attractive city images can be constructed (Molotch, Freudenburg & Paulsen 2000). As cities come into increasing and constant competition with one another, they look for any edge they can gain against one another, with authenticity, or factors that distinguish one city from another and make it unique, becoming a driving factor. It is in this congealed city character that city marketers search for an edge, both drawing from and reconstructing new histories of the city in question, activating certain populations' relations to the city and silencing others, constructing false but consumable authenticities (Zukin 2008, 2009, 2011). Ephemeral and recurrents events and festivals become new sources of authenticity that serve to also link a continuity in the city image of a city while at the same time open the city's image open to continued fluctuation (Yuan & Chong 2007).

City images fluctuate overtime from both the actions of city residents, as well as continual rebranding and highlighting of different histories, amenities, industries, and cultural offerings put forward by chambers of commerce and growth coalition leaders. The question becomes: Where do city images exist? How are they accesible? How do they change over time? What do people look for in cities? How does this determine what types of city images are being constructed? Through what mediums are they produced? At what level do they operate? A few studies have attempted to capture and meaure city image through interviews, looking at what residents and tourists look for and define as important to them, and using these measures in following surveys to measure which of these factors is present in different cities (Gliboa et al 2015). These studies conduct market research on those they see as the consumers of the city to determine what they look for in cities, as well as how they currently view any given city that the surveys are conducted in. These studies do not lend themselves well to studying city images across time, in accessing congealed examples of city image that are locked away in media representations, texts, actual marketing materials, and histories inaccessible by survey research. 

It is my project to look for new ways to begin to study where city images are being produced and crafted, and looking at how these city images are circulated. DiMaggio et al (2013) define topic modeling algorithms as "a suite of machine learning methods for discovering hidden thematic structure in large collections of documents. With a collection of documents as input, a topic model can produce a set of interpretable ‘‘topics’’ (i.e., groups of words that are associated under a single theme) and assess the strength with which each document exhibits those topics" (Pg.577). I will be using the specific topic modeling method of Latent Dirichlet Allocation (LDA), producing topics or themes. Dimaggio et al describe a topic as being "formally defined as a distribution over a vocabulary. Terms that are prominent within a topic are those that tend to occur in documents together more frequently than one would expect by chance" (pg.578). 

The number of topics pulled from the corpus for analysis is something determined ahead of time. There is no real rule for determining how many topics to set in advance, the general rule being to run multiple topic models with different numbers of topics pre-specificed and choosing that model whose topics are the most interpretable and understandable by the researcher and their prior knolwedge of the subject area. Being that topic models are inductively derived, it is hard for most reserchers to be able to determine ahead of time what they want to be looking for. In my case, however, I know one of the topics that I want to see emerging from the data: the city image. I will be operationalizing and defining the concept of the city image by looking for a topic which has Boston unique location and place based information, providing words that cannot be associated with any other city. Being that my listing data is bound to contain city-specific information, ehing the nature of the commodity being sold, it is reasonable to assume that a topic corresponding to this operationalization of city image should emerge from the data. My measure and rule of thumb for determining whether the number of topics chosen is good or not will rely on when I see this city image topic(s) emerging clearly and crystallizing. When I see this, it becomes more a task of seeing whether the other topics presented are legible. If they are, this will serve as my topic model.

### 1. Fit a Topic Model, using LDA
Now we're ready to fit the model. This requires the use of CountVecorizer, which we've already used, and the scikit-learn function LatentDirichletAllocation.

See here for more information about this function.

```python
#Here we have our functions for Topic Modeling and LDA that we will use in our analysis
#I ran the topic model a few times with different numbers of topics, and settled on 6 topics as this number of topics 
#produced the differentiation of spatial scale description I was looking for, with two potential topics roughly 
#interpretable as city image


####Adopted From: 
#Author: Olivier Grisel <olivier.grisel@ensta.org>
#         Lars Buitinck
#         Chyi-Kwei Yau <chyikwei.yau@gmail.com>
# License: BSD 3 clause

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

n_samples = 2000
n_topics = 6
n_top_words = 50

##This is a function to print out the top words for each topic in a pretty way.
#Don't worry too much about understanding every line of this code.
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
    print()
```

```python
# Vectorize our text using CountVectorizer
print("Extracting tf features for LDA...")
tf_vectorizer = CountVectorizer(max_df=0.80, min_df=50,
                                max_features=None,
                                stop_words='english'
                                )

tf = tf_vectorizer.fit_transform(df_abnb.listing)
```

```python
print("Fitting LDA models with tf features, "
      "n_samples=%d and n_topics=%d..."
      % (n_samples, n_topics))

#define the lda function, with desired options
#Check the documentation, linked above, to look through the options
lda = LatentDirichletAllocation(n_topics=n_topics, max_iter=20,
                                learning_method='online',
                                learning_offset=80.,
                                total_samples=n_samples,
                                random_state=0)
#fit the model
lda.fit(tf)
```

```python
#print the top words per topic, using the function defined above.
#Unlike R, which has a built-in function to print top words, we have to write our own for scikit-learn
#I think this demonstrates the different aims of the two packages: R is for social scientists, Python for computer scientists

print("\nTopics in LDA model:")
tf_feature_names = tf_vectorizer.get_feature_names()
print_top_words(lda, tf_feature_names, n_top_words)
```

### 2. Document by Topic Distribution
We want to find the most representative texts for each topic, to see which listings invoke these topics the most, so then we can group these listings by neighbourhood as established above. A simple way to do this (but not memory efficient), is to merge the topic distribution back into the Pandas dataframe.

First get the topic distribution array.

```python
#Creating measures to determine topic-loading per listing

topic_dist = lda.transform(tf)
topic_dist
```

```python
#adding topic loading measures to the dataframe

topic_dist_df = pd.DataFrame(topic_dist)
df_w_topics = topic_dist_df.join(df_abnb)
df_w_topics
```

I want to move on to the next stage of analysis, where I investigate how the presence of these topics in a listing is influenced by the geographic location of the unit. I will be looking at how different marketing information is required or more successful in some locations than others, this leading to highlighting of different place-based information. Here is where I will ask where the city-image, in terms of that topic, is most being marketed.

I will do so by grouping the listings by neighbourhood and averaging the topic loadings for each neighbourhood to see how geographic location effects which topics are chosen or available for marketing use.

```python
topic_columns = range(0,6)
for num in topic_columns:
    print(num)
```

```python
###Grouping topic loadings by Neighbourhood

grouped = df_w_topics.groupby('neighbourhood_y')
for num in topic_columns:
    print(num)
    print(grouped[num].mean())
```

The Topic 1: City Image 1, corresponding to the Creative Class Type of City Image, is most present in Allston, Brighton, Dorchester, the Longwood Medical Area, and Roxbury, all sites of universities, medical centers, and large employment institutions. 

The Topic 2: City Image 2, corresponding to the City Character/Authenticity Type of City Image, is most Present in Beacon Hill, the North End, Back Bay, Bay Village, the South End, Downtown, and Charlestown. These are all sites of heavy tourism, as well as high levels of consumption in the form of shopping and restaurant districts, all also areas of either high affluence or neo-bohemian (Lloyd 2002) sites of art and culture.   

#### Graphing the Geographic Distribution

Six Bar Plots of the average invoking of each topic by neighbourhood.

```python
import matplotlib.pyplot as plt
fig1 = plt.figure(figsize=(30,20))
chrt = 0
for num in topic_columns:
    chrt += 1 
    ax = fig1.add_subplot(2,3, chrt)
    grouped[num].mean().plot(kind = 'bar', yerr = grouped[num].std(), ylim=0, ax=ax, title=num)

fig1.tight_layout()
plt.show()
```

### 3. Words Aligned with each Topic
Following DiMaggio et al., we can calculate the total number of words aligned with each topic, and compare by neighbourhood, using this slightly different methodology of topic measurement to test the findings I found above.

```python
#first create word count column

df_w_topics['word_count'] = df_w_topics['listing'].apply(lambda x: len(str(x).split()))
df_w_topics['word_count']
```

```python
col_list = []
for num in topic_columns:
    col = "%d_wc" % num
    col_list.append(col)
    #Solution
    df_w_topics[col] = df_w_topics[num] * df_w_topics['word_count']
    
df_w_topics
```

```python
#EX: What is the total number of words aligned with each topic, by neighbourhood?
#EX: What is the proportion of total words aligned with each topic, by neighbourhood?

grouped = df_w_topics.groupby('neighbourhood_y')
for e in col_list:
    print(e)
    print(grouped[e].sum()/grouped['word_count'].sum())
```

The findings are largely the same as those arrived at above, with each City Image Topic being invoked the same interms of word count as they were by prevalence of topic loadings in the analysis above.

We'll plot the findings again for quick visual confirmation. 

```python
fig2 = plt.figure(figsize=(20,10))
chrt = 0
for e in col_list:
    chrt += 1 
    ax2 = fig2.add_subplot(2,3, chrt)
    (grouped[e].sum()/grouped['word_count'].sum()).plot(kind = 'bar', ylim=0, ax=ax2, title=e)

fig2.tight_layout()
plt.show()
```
