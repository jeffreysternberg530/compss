### Exercise 1.1: Nietzsche's Beyond Good and Evil

```python
#Exericse 1.1 

text = open('../data/beyond_good_and_evil.txt', encoding = 'utf-8').read()

#For punctuation use the list from the string library
import string
punct_list = list(string.punctuation)

#Function to Tokenize Text
def word_tokenize(text):
    text = text.lower()
    text_clean = ''.join([e for e in text if e not in punct_list])
    text_token =  text_clean.split()
    return text_token

#Tokenizing Text
text_tokens = word_tokenize(text)
text_tokens
```

```python
#Counting Words
counts = dict()
for word in text_tokens:
    if word not in counts:
        counts[word] = 1
    else:
        counts[word] += 1
counts
```

```python
#Frequency of Words
counts.items()

freq_words = []
for key, val in counts.items():
    freq_words.append((val, key))

freq_words.sort(reverse=True)
freq_words
```
